## Create Dataset using yfinance for :
- ADRO.JK (Energy Company)
- BBCA.JK (Banking)
- TLKM.JK (Telecommunication)

```python
# Create Dataset

import yfinance as yf
import pandas as pd
from datetime import datetime, timedelta

# Three Indonesia Corp which different sectors listed in Yahoo Finance
ticker_symbols = ['ADRO.JK', 'BBCA.JK', 'TLKM.JK']

# Define start and end dates
# START_DATE = (datetime.today() - timedelta(days=100)).strftime('%Y-%m-%d')
# CURRENT_DATE = datetime.today().strftime('%Y-%m-%d')
START_DATE = "2020-01-01"
END_DATE = datetime.today().strftime('%Y-%m-%d')

# Join ticker symbols into a single string separated by comma
tickers = ",".join(ticker_symbols)

# Download data from Yahoo Finance
try:
    data = yf.download(tickers, start=START_DATE, end=END_DATE, group_by='ticker', interval='1d')
except yf.exceptions.YFTzMissingError as e:
    print(f"Failed downloads: {e.args[0]}")
    # Handle invalid tickers here, e.g., remove them from the list and retry

# Creaate DataFrame
data = pd.DataFrame(data)
data
```

## EDA
- Closing Price
- Close Price and 30-Day Moving Average
- Correlation between Stocks
- Daily Return
- Monthly and Yearly Average Closing Price

```python
# EDA

import matplotlib.pyplot as plt

# Plot closing prices for each ticker
for ticker in ticker_symbols:
    plt.figure(figsize=(12, 6))
    data[ticker]['Close'].plot(title=f'{ticker} Closing Prices')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.show()
```

```python
# Create a dictionary to hold the data with moving averages
ma_data = {}

for ticker in ticker_symbols:
    # Extract close prices for the ticker and calculate the 30-day moving average
    close_prices = data[ticker]['Close']
    ma_data[ticker] = pd.DataFrame({
        'Close': close_prices,
        'MA30': close_prices.rolling(window=30).mean()
    })

    # Plot close prices and 30-day moving average
    plt.figure(figsize=(12, 6))
    plt.plot(ma_data[ticker]['Close'], label='Closing Price')
    plt.plot(ma_data[ticker]['MA30'], label='30-Day Moving Average')
    plt.title(f'{ticker} - Close Price and 30-Day Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.show()
```

```python
# Extract closing prices of all tickers
close_prices = pd.DataFrame({ticker: data[ticker]['Close'] for ticker in ticker_symbols})

# Calculate correlation matrix
correlation_matrix = close_prices.corr()
print("Correlation Matrix:\n", correlation_matrix)

# Heatmap of the correlation matrix
import seaborn as sns
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation between Stocks')
plt.show()
```

```python
# Calculate daily returns
returns = close_prices.pct_change()

# Plot the daily returns
for ticker in ticker_symbols:
    plt.figure(figsize=(12, 6))
    returns[ticker].plot(title=f'{ticker} Daily Returns')
    plt.xlabel('Date')
    plt.ylabel('Daily Return')
    plt.show()

# Calculate volatility (standard deviation of returns)
volatility = returns.std()
print("Volatility:\n")
print(volatility)
```

```python
# Resample to monthly mean
monthly_prices = close_prices.resample('ME').mean()
monthly_prices.plot(figsize=(14, 7), title='Monthly Average Closing Prices')
plt.xlabel('Date')
plt.ylabel('Monthly Average Price')
plt.show()

# Resample to yearly mean
yearly_prices = close_prices.resample('YE').mean()
yearly_prices.plot(figsize=(14, 7), title='Yearly Average Closing Prices')
plt.xlabel('Year')
plt.ylabel('Yearly Average Price')
plt.show()
```

## Targets and Features Engineering

```python
# All columns

print(data.columns)
# Therea are 2 columns for indexes: Ticker & Price
```

```python
for ticker in ticker_symbols:
    ticker_feature_canditates = [col for col in data.columns if ticker in col and 'target' not in col]    
    print(ticker_feature_canditates)
```

```python
# Define columns format 
data.columns = ['_'.join(col).replace(' ', '') for col in data.columns]
print(data.columns)
```

```python
# Features and Target Engineering
# Add 1-day Moving Average (MA1) for each ticker
data = data.copy()  # Ensure data is a full copy, not a view

for ticker in ticker_symbols:
    close_col = f"{ticker}_Close"
    ma_col = f"{ticker}_MA1"
    data[ma_col] = data[close_col].rolling(window=1).mean()

print(data.columns)
```

```python
# Features and Target Engineering
# Add 30-day Moving Average (MA30) for each ticker
data = data.copy()  # Ensure data is a full copy, not a view

for ticker in ticker_symbols:
    close_col = f"{ticker}_Close"
    ma_col = f"{ticker}_MA30"
    data[ma_col] = data[close_col].rolling(window=30).mean()

print(data.columns)
```

```python
# Define Targets and Features for every Ticker

ticker_symbols = ['ADRO.JK', 'BBCA.JK', 'TLKM.JK']
targets = ['UpdDown', 'DailyResult', 'Votality']

for ticker in ticker_symbols:
    for target in targets:
        match target:
            case 'UpdDown':
                features = f'{ticker}_Open, {ticker}_Close, {ticker}_MA1'
            case 'DailyResult':
                features = f'{ticker}_Open, {ticker}_Close, {ticker}_High, {ticker}_Low'
            case 'Votality':
                features = f'{ticker}_Open, {ticker}_Close, {ticker}_Volume, {ticker}_MA30'
            case _:
                features = "Unknown target"
        
        print(f'\nFeatures for {ticker} with target {target}:')
        print(features)
```
